```python
# Read a word from user and print all synonyms  of that word
word = input("Enter a word : ")
synonyms = set()
for w in wordnet.synsets(word):
    if w.lemmas()[0].name() != word:
        synonyms.add(w.lemmas()[0].name())
print(synonyms)
```

```python
# Read a word from user and print all noun meanings from it
word = input("Enter a word : ")

for w in wordnet.synsets(word):
    if  w.pos() == 'n':
        print(w.name(),"-->",w.definition())
```

```python
# Read a word from user and print all noun meanings from it
word = input("Enter a word : ")

for w in wordnet.synsets(word,pos="n"):
    print(w.name(),"-->",w.definition())
```

```python
sent = """The fiber is most often spun into yarn or thread and used to make a soft, breathable, and durable textile"""

# Find the synonmys of all the adjectives in the above sentence
for word in sent.split():
    for w in wordnet.synsets(word):
        if w.pos() == 'a':
            print(w.lemmas()[0].name())
```

```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag
```

```python
adj = [word[0] for word in pos_tag(word_tokenize(sent)) if word[1].startswith('JJ')]
adj
```

```python
for word in adj:
    syns = set()
    for syn in wordnet.synsets(word):
        syns.add(syn.lemmas()[0].name())
    syns.discard(word)
    print("synonyms of ",word,"are",syns)
```

```python
ant = set()
for synset in wordnet.synsets("run"):
    for lemma in synset.lemmas():
        if lemma.antonyms():
            ant.add(lemma.antonyms()[0].name())
print(ant)
```
