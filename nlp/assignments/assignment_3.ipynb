```python
# Read a word from user and print all meanings of it using wordnet
from nltk.corpus import wordnet
word = input("Enter a word: ")
syns = wordnet.synsets(word)
print("Meanings of the word: ")
for syn in syns:
    print(syn.definition())
```

```python
# Read a word from user and print all noun meanings of it.
from nltk.corpus import wordnet
word = input("Enter a word: ")
syns = wordnet.synsets(word)
print("Noun meanings of the word: ")
for syn in syns:
    if syn.pos() == 'n':
        print(syn.definition())
```

```python
from nltk import pos_tag
from nltk.tokenize import word_tokenize
from nltk import  RegexpParser

# 1. Apply chuncking and show the chunks of type <VBD><DT>
grammar = "NP: {<VBD><DT>}"
cp = RegexpParser(grammar)
tagged = pos_tag(word_tokenize(text))

result = cp.parse(tagged)
result
```

```python
# 2. Find all the named entities present
import spacy
from spacy import displacy

nlp = spacy.load('en_core_web_sm')
doc = nlp(text)
# for ent in doc.ents:
#     print(ent.text, ent.label_)
displacy.render(doc, style='ent', jupyter=True)
```

```python
# 3. Access all the verbs and apply the porter stemmer and lancaster stemmer
from nltk.stem import PorterStemmer, LancasterStemmer
porter = PorterStemmer()
lancaster = LancasterStemmer()
verbs = [word for word, pos in tagged if pos.startswith('V') and word.isalpha()]
print("Porter Stemmer: ")
for verb in verbs:
    print(porter.stem(verb))
print("Lancaster Stemmer: ")
for verb in verbs:
    print(lancaster.stem(verb))
```

```python
# 4. Access all the verbs in past tense and lemmatize them. Show only unique verbs.
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
verbs_past = [word for word, pos in tagged if pos.startswith('VBD') and word.isalpha()]
print("Lemmatized verbs: ")
lemmatized_verbs = [lemmatizer.lemmatize(verb) for verb in verbs_past]
print(set(lemmatized_verbs))
```

```python
# 5. Read only the adjectives from the text and find their synonyms.
from nltk.corpus import wordnet

adjectives = [word for word, pos in tagged if pos.startswith('JJ') and word.isalpha()]
print("Adjectives: ")
print(adjectives)
print("Synonyms: ")

for adjective in adjectives:
    synsets = wordnet.synsets(adjective, pos=wordnet.ADJ)  # Get only adjective synsets
    synonyms = []
    for synset in synsets:
        for lemma in synset.lemmas():
            synonyms.append(lemma.name())
    print(f"Synonyms for {adjective}: {synonyms}")
```

```python
# 6. Read all the verbs and print their antonyms.
from nltk.corpus import wordnet

verbs = [word for word, pos in tagged if pos.startswith('V') and word.isalpha()]

print("Antonyms: ")
for verb in verbs:
    antonyms = []
    synsets = wordnet.synsets(verb, pos=wordnet.VERB)  # Get only verb synsets
    for synset in synsets:
        for lemma in synset.lemmas():
            if lemma.antonyms():
                antonyms.append(lemma.antonyms()[0].name())
    print(f"Antonyms for {verb}: {antonyms}")
```

```python
# 7. Print all the entities marked as DATE and TIME in the data
import spacy
from spacy import displacy

nlp = spacy.load('en_core_web_sm')
doc = nlp(text)
for ent in doc.ents:
    if ent.label_ == 'DATE' or ent.label_ == 'TIME':
        print(ent.text, ent.label_)
```

```python
# 8. Print all the named of locations present in the article.
import spacy
from spacy import displacy

nlp = spacy.load('en_core_web_sm')
doc = nlp(text)
for ent in doc.ents:
    if ent.label_ == 'GPE':
        print(ent.text, ent.label_)
```
