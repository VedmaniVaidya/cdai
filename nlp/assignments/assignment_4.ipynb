```python
# Open the wikipedia page https://en.wikipedia.org/wiki/Saturn
# 1. Scrap the web and find top 10 nouns from the page
# 2. Print all the unique verbs in their root form.
# 3. Extract all the <DT><JJ><NN> phrases from the article.
# 4. Summarize the text using TextRank Algorithm
# 5. Summarize the text using extractive summarization
```

```python
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import word_tokenize
from nltk import FreqDist


headers = {'User-Agent': 'Mozilla/5.0'}
url = 'https://en.wikipedia.org/wiki/Saturn'
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')
text = soup.get_text()

tokens = word_tokenize(text)
#remove stopwords and punctuations
stopwords = nltk.corpus.stopwords.words('english')

tokens = [token for token in tokens if token.isalpha() and token not in stopwords]
tagged = nltk.pos_tag(tokens)

nouns = [word for word,pos in tagged if pos == 'NN']

#find top 10 nouns
fdist = FreqDist(nouns)
print(fdist.most_common(10))
```

```python
# 2. Print all the unique verbs in their root form.
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
verbs = [word for word,pos in tagged if pos == 'VB']
lemmatized_verbs = [lemmatizer.lemmatize(verb) for verb in verbs]
print(set(lemmatized_verbs))
```

```python
# 3. Extract all the <DT><JJ><NN> phrases from the article.
from nltk.corpus import wordnet as wn
from nltk import RegexpParser

grammar = "NP: {<DT>?<JJ>*<NN>}"
cp = nltk.RegexpParser(grammar)
result = cp.parse(tagged)

result
```

```python
# 4. Summarize the text using TextRank Algorithm
#Import the TextRank summarizer
from sumy.summarizers.text_rank import TextRankSummarizer

#Importing the parser and tokenizer
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer

#Creating a parser object
parser=PlaintextParser.from_string(text,Tokenizer("english"))
summarizer=TextRankSummarizer()
summary=summarizer(parser.document,3)

#Printing the summary
for sentence in summary:
    print(sentence)
```

```python
headers = {'User-Agent': 'Mozilla/5.0'}
url = 'https://en.wikipedia.org/wiki/Saturn'
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')
text = soup.get_text()
```

```python
# 5. Summarize the text using extractive summarization
from sumy.summarizers.lex_rank import LexRankSummarizer
lex_rank_summarizer = LexRankSummarizer()
parser=PlaintextParser.from_string(text,Tokenizer("english"))
summary = lex_rank_summarizer(parser.document,sentences_count=3)
for sentence in summary:
    print(sentence)
```

```python
from sumy.summarizers.lsa import LsaSummarizer
lsa_summarizer = LsaSummarizer()
parser=PlaintextParser.from_string(text,Tokenizer("english"))
summary = lsa_summarizer(parser.document,sentences_count=3)
for sentence in summary:
    print(sentence)
```
