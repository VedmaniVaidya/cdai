```python
from train import k_fold_train
```

```python
loss, metrics = k_fold_train()
```

```python
#plot training loss
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
sns.set_palette('Set2')
plt.figure(figsize=(12,8))

for id, fold_loss in enumerate(loss):
    df = pd.DataFrame(fold_loss)
    plt.plot(df, label=f'fold_{id}')

plt.title('loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.xticks(range(0, len(df), 1))
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
sns.set_palette('Set2')
plt.figure(figsize=(12,8))

for id, fold_metric in enumerate(metrics):
    df = pd.DataFrame(fold_metric)
    plt.plot(df['loss'], label=f'fold_{id}_metric')

plt.title('Eval_loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.xticks(range(0, len(df), 1))
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
sns.set_palette('Set2')
plt.figure(figsize=(12,8))

for id, fold_metric in enumerate(metrics):
    df = pd.DataFrame(fold_metric)
    plt.plot(df['rocauc'], label=f'fold_{id}_metric')

plt.title('ROC_AUC')
plt.xlabel('epoch')
plt.ylabel('rocauc')
plt.xticks(range(0, len(df), 1))
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
sns.set_palette('Set2')
plt.figure(figsize=(12,8))

for id, fold_metric in enumerate(metrics):
    df = pd.DataFrame(fold_metric)
    plt.plot(df['accuracy'], label=f'fold_{id}_metric')

plt.title('Eval Accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.xticks(range(0, len(df), 1))
plt.legend()
plt.show()
```
