```python
import pandas as pd
from pathlib import Path
import shutil
```

```python
#these dataframes are created using ImageFolder from pytorch to get paths and labels then using sklearn to split into train and validation with seed=42 and shuffle=True
train_df = pd.read_csv('data/train.csv')
validation_df = pd.read_csv('data/validation.csv')
```

```python
train_df
```

```python
validation_df
```

```python
root_dir = Path('data')
```

```python
for label in train_df['label'].unique().tolist():
    root_dir.joinpath('train', label).mkdir(parents=True, exist_ok=True)
for label in validation_df['label'].unique().tolist():
    root_dir.joinpath('validation', label).mkdir(parents=True, exist_ok=True)
```

```python
#copy images to train folder
for path in train_df['image_path']:
    image_path = Path(path)
    label = image_path.parent.name
    image_name = image_path.name
    shutil.copy(path, root_dir.joinpath('train', label, image_name))
```

```python
#copy images to validation folder
for path in validation_df['image_path']:
    image_path = Path(path)
    label = image_path.parent.name
    image_name = image_path.name
    shutil.copy(path, root_dir.joinpath('validation', label, image_name))
```
